```python
# import dependencies: matplotlib along with pyplot
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
# create the plot iwth ax.plt()
# fig, ax = plt.subplots()
# ax.plot(x_axis, y_axis)
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# the plt.show function
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# create plot and add label for legend
plt.plot(x_axis, y_axis, marker='*', color='blue', linewidth=2, label='Boston')

# create labes for the x and y axes
plt.xlabel('Date')
plt.ylabel('Fare($)')

# set the y limits to 0 and 45
plt.ylim(0, 45)

# create title
plt.title("PyBer Fare by Month")

# add a grid
plt.grid()

# add legend
plt.legend()
```

```python
### annotations using ooi
#Matplotlib Object-Oriented Functions	Feature
#fig, ax = plt.subplots(figsize=(w, h))	Change the size of the figure in pixels. Add this in the subplots() function.
#ax.plot(x, y, label='line')	Add a label that will be added to the legend.
#ax.set_ylim(min, max)	Sets the min and max range of the y-axis.
#ax.set_xlim(min, max)	Sets the min and max range of the x-axis.
#ax.set_xlabel('x label')	Add a label to the x-axis.
#ax.set_ylabel('y label')	Add a label to the y-axis.
#ax.set_title("Title")	Add a title.
#ax.legend()	Add a legend.
#ax.grid()	Add a grid to the chart.
#** plt.savefig("add a path and figure extension")	Saves the figure with the given extension. Added at the end of your script.
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker='D', color='green', linewidth=2, label='Boston')
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_ylim(0, 45)
ax.grid()
ax.legend()
ax.set_title('PyBer Fare by Month')
```

```python
# create vertical bar chart
plt.bar(x_axis, y_axis)
```

```python
# annotate the bar chart
plt.bar(x_axis, y_axis, color='green', label='Boston')

# create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# create title
plt.title("PyBer Fare by Month")

# add legend
plt.legend()
```

```python
# create horizontal bar chart
plt.barh(x_axis, y_axis, color='magenta', label='Boston')

# label axes
plt.xlabel('Fare($)')
plt.ylabel('Date')

# add legend
plt.legend()

# invert y_axis
plt.gca().invert_yaxis()
```

```python
# create vertical bar using ooi method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='black', label='New York')
```

```python
#create horizontal bar chart using ooi method
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color='cyan', label='Chicago')

# label axes
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')

# add legend
ax.legend()

# invert y axis
fig.gca().invert_yaxis()
```

```python
# create scatter plot using MATLAB approach
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]

plt.scatter(x, y, c='green', marker='x', s=x)
plt.xlabel('Total Number of Rides(Per City)')
plt.ylabel('Average Fare')
plt.title('Average Fare vs Total Rides')
```

```python
# create scatter plot: 'o' changes .plot from line to scatter 
plt.plot(y_axis, x_axis, 'o', c='red', label='Chicago')
plt.legend()
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
# make a bubble chart using MATLAB
##y_axis_larger = []
##for data in y_axis:
    #y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=[i*3 for i in y_axis])
```

```python
# make scatter plot using ooi method
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# make bubble chart using ooi approach
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c='skyblue', edgecolor='black', alpha=0.8, linewidths=2, label='Boston')
ax.set_title('PyBer Fare by Month')
ax.legend()
fig.gca().invert_yaxis()
ax.set_xlim(0, 50)
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
```

```python
# create pie chart using MATLAB
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%', 
    labeldistance=.8,)

plt.show()
```

```python
# create pie chart using ooi method
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,labels=x_axis, autopct='%.1f%%', explode=explode_values, colors=colors, shadow=True, startangle=100, counterclock=False)
plt.show()
```
